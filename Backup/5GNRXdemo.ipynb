```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from NRX import Transmitter

showfigure = True

transmit = Transmitter(
    num_rx=1,
    num_tx=1,
    batch_size=64,
    fft_size=76,
    num_ofdm_symbols=14,
    num_bits_per_symbol=4,
    subcarrier_spacing=60e3,
    USE_LDPC=False,
    pilot_pattern="kronecker",
    guards=True,
    showfig=showfigure,
)  # "kronecker" "empty"
```

```python
# Test the transmitter
b_hat, BER = transmit(ebno_db=5.0, channeltype="ofdm", perfect_csi=False)

# Print the results
print("Estimated bits:", b_hat)
print("Bit Error Rate (BER):", BER)
```
